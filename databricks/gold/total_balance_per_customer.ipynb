```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
accounts_df = spark.read.format("csv")\
    .option("header",True)\
        .option("inferSchema",True)\
            .load("s3://end-to-end-banking-data-pipeline/silver/cleaned_accounts")
accounts_df.display()
```

```python
# 1. Calculate total balance per customer
customer_balance_df = accounts_df.groupBy("customer_id").agg(
    {"balance": "sum"}
).withColumnRenamed("sum(balance)", "total_balance")
```

```python
# load to s3 gold layer
customer_balance_df.write.mode("overwrite").format("csv")\
    .option("header", "true")\
    .save("s3://end-to-end-banking-data-pipeline/gold/total_balance_per_customer/")
customer_balance_df.write.mode("overwrite").saveAsTable("workspace.banking_data.total_balance_per_customer")
```
